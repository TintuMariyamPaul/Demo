```python
#Test that journeys within half an hour (1800 seconds) of a previous journey are from the destination of the first journey
class BusRoute:
  def __init__(self, origin, destination, duration):
        self.origin = origin
        self.destination = destination
        self.duration = duration

  def __str__(self):
    return f"{self.origin} to {self.destination} in {self.duration} seconds"
  def __repr__(self):
    return f"BusRoute({self.origin}, {self.destination}, {self.duration})"

class Journey:
  def __init__(self, route, busID, startTime):
    self.route = route
    self.busID = busID
    self.startTime = startTime
  def __str__(self):
    return f"{self.route} on bus {self.busID} at {self.startTime}"
  def __repr__(self):
    return f"Journey({self.route}, {self.busID}, {self.startTime})"

  def get_end_time(self):
    return self.startTime + self.route.duration

class Schedule:
    def __init__(self):
        self.routes = []
        self.journeys = []

    def add_route(self, route):
        if not isinstance(route, BusRoute):
            raise TypeError("Route must be instance of BusRoute")
        self.routes.append(route)


    def add_journey(self, journey):
      if not isinstance(journey, Journey):
        raise TypeError("Journey must be instance of Journey")


      for existing_journey in self.journeys:
        if existing_journey.busID == journey.busID:
          raise ValueError("Bus already scheduled for previous journey")
          if existing_journey.get_end_time() > journey.startTime:
            raise ValueError(f"bus {journey.busID}is scheduled for previous journey")
      for existing_journey in self.journeys:
        time_difference = journey.startTime - existing_journey.get_end_time()
        if 0 <=time_difference <= 1800:
          if existing_journey.route.origin != journey.route.origin:
            raise ValueError(f"Journey starting at {journey.startTime} must originate from "
                    f"{existing_journey.route.destination}, the destination of the previous journey"
                )
      self.journeys.append(journey)
    def get_routes(self):
      return self.routes
    def get_journeys(self):
      return self.journeys

import unittest
class TestSchedule(unittest.TestCase):
  def setUp(self):
    self.route1 = BusRoute("London", "Manchester", 1800)
    self.route2 = BusRoute("Manchester", "London", 1800)
    self.schedule = Schedule()
    self.schedule.add_route(self.route1)
    self.schedule.add_route(self.route2)
  def test_duplicate_bus_journey(self):
    journey1 = Journey(self.route1, "ABC123", 0)
    self.schedule.add_journey(journey1)
    journey2 = Journey(self.route2, "ABC123", 120)
    with self.assertRaises(ValueError) as context:
      self.schedule.add_journey(journey2)
    self.assertEqual(str(context.exception), "Bus already scheduled for previous journey")
  def test_valid_journey(self):
    journey1 = Journey(self.route1, "ABC123", 0)
    self.schedule.add_journey(journey1)
    journey2 = Journey(self.route2, "DEF456", 120)

    try:
      self.schedule.add_journey(journey2)
    except ValueError as e:
      self.fail(f"Unexpected ValueError: {e}")
  def test_duplicate_bus_journey_within_half_hour(self):
    journey1 = Journey(self.route1, "ABC123", 0)
    self.schedule.add_journey(journey1)
    journey2 = Journey(self.route2, "ABC123", 7200)
    with self.assertRaises(ValueError) as context:
      self.schedule.add_journey(journey2)
      self.assertEqual(str(context.exception), "bus ABC123 is scheduled for previous journey")

def t():
    unittest.main(argv=['ignored'], exit=False)

t()
```
